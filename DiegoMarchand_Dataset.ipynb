```python
"""
Este Dataset es presenta data de entrenamiento para detectar fraude en los prestamos bancarios
Las columnas presentadas son:

ID_PRESTAMO	ID del préstamo realizado
CL_MOROSO	Variable objetivo (1 - cliente con dificultades de pago: se atrasó más de X días en al menos una de las primeras Y cuotas del préstamo de nuestra muestra, 0 - resto de casos)
T_CONTRATO	Identificación del préstamo en efectivo o renovables
T_SEXO	Sexo del cliente
CARRO_PROPIO	Flag si el cliente cuenta con auto
INMUEBLE_PROPIO	Flag si cliente cuenta con vivienda propia
N_HIJOS	Número de hijos del cliente
INGRESOS_ANUALES	Ingresos del cliente
P_CREDITO	Monto del crédito del préstamo
P_ANUALIDAD	Anualidad del préstamo
PRECIO_BIEN_ADQ_CL	Para préstamos de consumo es el precio de los bienes por los cuales se otorga el préstamo.
T_ACOMPAÑANTE	Quién acompañaba al cliente cuando estaba solicitando el préstamo
F_PRINCIPAL_INGRESOS	Fuente principal de ingreso de los clientes (empresario, trabajador, baja por maternidad,…)
NIVEL_EDUCACION	Nivel de educación más alto que alcanzó el cliente
ESTADO_CIVIL	Estado civil del cliente
T_VIVIENDA	Cuál es la situación de vivienda del cliente (alquilando, viviendo con los padres,...)
POBLACION_REGION_RELATIVA	Población normalizada de la región donde vive el cliente (un número mayor significa que el cliente vive en una región más poblada) - Elaborado por investigador
EDAD_DIAS	Edad del cliente en días al momento de la solicitud
TIEMPO_EMPLEO_DIAS	Cuántos días antes de la solicitud la persona comenzó el empleo actual
REGISTRO_CAMBIADO_DIAS	¿Cuántos días antes de la aplicación el cliente cambió su registro?
DNI_CAMBIADO_DIAS	¿Cuántos días antes de la solicitud el cliente cambió el documento de identidad con el que solicitó el préstamo?
AUTO_PROPIO_AÑOS	Antiguedad del coche del cliente
FLAG_CELULAR	El cliente proporcionó teléfono móvil (1=SÍ, 0=NO)
FLAG_TELEFONO_HOGAR	¿El cliente proporcionó el teléfono del trabajo (1 = SÍ, 0 = NO)
FLAG_TELEFONO_TRABAJO	¿El cliente proporcionó el teléfono de su casa? (1=SÍ, 0=NO)
FLAG_CELULAR_SEC	¿Era accesible el teléfono móvil? (1=SÍ, 0=NO)
FLAG_EMAIL	¿El cliente proporcionó correo electrónico (1 = SÍ, 0 = NO)
T_PUESTO_TRABAJO	Qué tipo de ocupación tiene el cliente
N_FAMILIARES	¿Cuántos miembros de la familia tiene el cliente?
CL_REGION_RATING	Rating de la región donde vive el cliente (1,2,3) - Rating elaborado por investigador
VL_TRABAJO_REGION_RATING	Rating de la región donde vive el cliente teniendo en cuenta la ciudad (1,2,3) - Rating elaborado por investigador
FECHA_SOLICITUD	¿En qué día de la semana solicitó el préstamo el cliente?
HORA_SOLICITUD	Aproximadamente a qué hora solicitó el préstamo el cliente
REG_REGION_NO_DNI_REGION	Marcar si la dirección registrada del cliente no coincide con la dirección de contacto en el DNI (1=diferente, 0=igual, a nivel de región)
REG_REGION_NO_REGION_TRABAJO	Marcar si la dirección registrada del cliente no coincide con la dirección del trabajo (1=diferente, 0=igual, a nivel de región)
DNI_REGION_NO_REGION_TRABAJO	Marcar si la dirección de contacto en el DNI del cliente no coincide con la dirección del trabajo (1=diferente, 0=igual, a nivel de región)
REG_CIUDAD_NO_DNI_CIUDAD	Marcar si la dirección registrada del cliente no coincide con la dirección de contacto en el DNI (1=diferente, 0=igual, a nivel de ciudad)
REG_CIUDAD_NO_CIUDAD_TRABAJO	Marcar si la dirección registrada del cliente no coincide con la dirección del trabajo (1=diferente, 0=igual, a nivel de ciudad)
DNI_CIUDAD_NO_CIUDAD_TRABAJO	Marcar si la dirección de contacto del cliente en el DNI no coincide con la dirección del trabajo (1=diferente, 0=igual, a nivel de ciudad)
T_ORGANIZACION_TRABAJO	Tipo de organización donde trabaja el cliente
OBS_30_DIAS_CIRCULO_SOCIAL	Cuantas observaciones del entorno social del cliente con incumplimiento observable de 30 DVP (días vencidos del prestamo)
INF_30_DIAS_CIRCULO_SOCIAL	Cuantas observaciones del entorno social del cliente incumplidas en 30 DVP (días vencidos del prestamo)
OBS_60_DIAS_CIRCULO_SOCIAL	Cuantas observaciones del entorno social del cliente con incumplimiento observable de 60 DVP (días vencidos del prestamo)
INF_60_DIAS_CIRCULO_SOCIAL	Cuantas observaciones del entorno social del cliente incumplidas en 60 DVP (días vencidos del prestamo)
ULT_CAMBIO_TELEFONO_DIAS	¿Cuántos días antes de la aplicación el cliente cambió de teléfono?
FLAG_DOCUMENT_2	¿El cliente proporcionó el documento 2?

La base de datos ha sido obtenida de: "https://www.kaggle.com/datasets/mishra5001/credit-card"

"""
```
